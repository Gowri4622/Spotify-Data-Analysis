### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load Dataset

```python
tracks_df=pd.read_csv('tracks.csv')
tracks_df.head()
```

#### Checking Null values

```python
pd.isnull(tracks_df).sum()
```

```python
tracks_df.info()
```

### Getting top 10 least popularity songs

```python
sorted_df=tracks_df.sort_values('popularity',ascending=True).head(10)
sorted_df
```

```python
tracks_df.describe().transpose()
```

### Getting top 10 popularity songs

```python
most_popular_df=tracks_df.query('popularity>90',inplace=False).sort_values('popularity',ascending=False)
most_popular_df[:10]
```

```python
tracks_df.set_index("release_date",inplace=True)
tracks_df.index=pd.to_datetime(tracks_df.index)
tracks_df.head()
```

```python
tracks_df[["artists"]].iloc[27]
```

### Converting duration in milli seconds to seconds

```python
tracks_df["duration"]=tracks_df["duration_ms"].apply(lambda x: round(x/1000))
tracks_df.drop("duration_ms", inplace=True, axis=1)
```

```python
tracks_df.duration.head()
```

```python
corr_df=tracks_df.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0, cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df=tracks_df.sample(int(0.004*len(tracks_df)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="loudness", x="energy", color="c").set(title="Loudness vs Energy Correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="popularity", x="acousticness", color="r").set(title="Popularity vs Acousticness Correlation")
```

```python
tracks_df['dates']=tracks_df.index.get_level_values('release_date')
tracks_df.dates=pd.to_datetime(tracks_df.dates)
years=tracks_df.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5, kind="hist").set(title="Number of songs per year")
```

```python
total_dr=tracks_df.duration
fig_dims=(18,7)
fig, ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

## Analysing SpotifyFeatures dataset

```python
df_genre=pd.read_csv("SpotifyFeatures.csv")
```

```python
df_genre.head()
```

```python
plt.title("Duration of the Songs in Different Genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous=df_genre.sort_values("popularity", ascending=False).head(10)
sns.barplot(y='genre', x='popularity', data=famous).set(title="Top 5 genres by popularity")
```
